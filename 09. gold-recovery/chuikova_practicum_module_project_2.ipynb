**Расчёт эффективности** 

Вам нужно смоделировать процесс восстановления золота из золотосодержащей руды. 
Эффективность обогащения рассчитывается по формуле

$$Recovery = \frac{C * (F - T)}{F * (C - T)}\ *100\%\ $$
где:
* *C* — доля золота в концентрате после флотации/очистки;
* *F* — доля золота в сырье/концентрате до флотации/очистки;
* *T* — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

**Метрика качества**

Для решения задачи введём новую метрику качества — **sMAPE** (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

Метрика sMAPE вычисляется так:

$$sMAPE = \frac{1}{N} \sum^{n}_{i=1}\frac{|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2} * 100\%\ $$

Обозначения:

* ${y_i}$ – значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество.
* $\hat{y_i}$ – значение предсказания для объекта с порядковым номером i, например, в тестовой выборке.
* *N* – Количество объектов в выборке.
* $\sum^{n}_{i=1}$ – cуммирование по всем объектам выборки (i меняется от 1 до N).

Нужно спрогнозировать сразу две величины:
* эффективность обогащения чернового концентрата `rougher.output.recovery`;
* эффективность обогащения финального концентрата `final.output.recovery`.

Итоговая метрика складывается из двух величин:
* ${y_i}$ – значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество.

$$Итоговое\ sMAPE = \text{25% * sMAPE(roughter) + 75% * sMAPE(final)}$$

### Проверим, что эффективность обогащения рассчитана правильно.

Эффективность обогащения рассчитывается по формуле

$$Recovery = \frac{C * (F - T)}{F * (C - T)}\ *100\%\ $$
где:
* *C* — доля золота в концентрате после флотации/очистки (данные содержатся в столбце `'rougher.output.concentrate_au'`);
* *F* — доля золота в сырье/концентрате до флотации/очистки (данные содержатся в столбце `'rougher.input.feed_au'`);
* *T* — доля золота в отвальных хвостах после флотации/очистки (данные содержатся в столбце`'rougher.output.tail_au'`).


Рассчитаем эффективность обогащения на обучающей выборке и сравним результаты со значениями в столбце `'rougher.output.recovery'`. Для того, чтобы оставить исходный датасет без изменений, сохраним столбец `'rougher.output.recovery'` в отдельную таблицу и добавим в нее столбец `'rougher.output.recovery_check'` с рассчитанными по формуле значениями.

```python
recovery_check = data_train[['rougher.output.recovery']]

# для удобства чтения кода введем переменные С, F, T и подставим их в формулу

C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']

recovery_check['rougher.output.recovery_check'] = (C * (F - T)) / (F * (C - T))*100
```

```python
recovery_check.head()
```

```python
recovery_check.hist(bins=50, figsize=(12,7));
```

Похоже, значения в столбце `'rougher.output.recovery'` рассчитаны верно. Чтобы проверить это, найдем  MAE между расчётами в `'rougher.output.recovery_check'` и значением признака в столбце `'rougher.output.recovery'`.

Перед тем как осуществлять расчет MAE, проверим `'rougher.output.recovery'` и `'rougher.output.recovery_check'` на наличие пропусков.

```python
recovery_check.isna().sum()
```

Пропусков нет. Теперь вычислим MAE.

```python
mean_absolute_error(recovery_check['rougher.output.recovery'], 
                    recovery_check['rougher.output.recovery_check'])
```

Значение MAE между расчётами в `'rougher.output.recovery_check'` и значением признака в столбце `'rougher.output.recovery'` очень мало. Можно сделать вывод, что эффективность обогащения рассчитана правильно.

### Напишем функцию для вычисления итоговой sMAPE.

Напишем две функции: для вычисления sMAPE и итоговой sMAPE.

Напомним формулы, по которым вычисляется sMAPE и итоговая sMAPE.

$$sMAPE = \frac{1}{N} \sum^{n}_{i=1}\frac{|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2} * 100\%\ $$

$$Итоговое\ sMAPE = \text{25% * sMAPE(roughter) + 75% * sMAPE(final)}$$


Где:
* ${y_i}$ – значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество.
* $\hat{y_i}$ – значение предсказания для объекта с порядковым номером i, например, в тестовой выборке.
* *N* – Количество объектов в выборке.
* $\sum^{n}_{i=1}$ – cуммирование по всем объектам выборки (i меняется от 1 до N).



```python
def smape(target, predictions):
    error = 2*np.abs(target-predictions)
    scale = np.abs(target)+np.abs(predictions)
    return np.mean(error / scale) * 100 

smape_scorer = make_scorer(smape, greater_is_better=False)
```

Поскольку установлено `greater_is_better=False` напишем функцию, переводящую отрицательные значения в положительные, которую в дальнейшем для удобства будем применять к значениям метрики sMAPE

```python
def convert(mean_score):
    if mean_score < 0:
        return mean_score * -1
```

```python
def final_smape(smape_roughter, smape_final):
    return 0.25 * smape_roughter + 0.75 * smape_final
```
